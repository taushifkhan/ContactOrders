```python
import MySQLdb as mdb
import numpy as Np
import pandas as pd
```

```python
con = mdb.connect('localhost','root','biokhan','proLego_v1')
cur = con.cursor()
```

```python
fl = open('./d1_cOS','r').readlines()
```

```python
d1_Data = []
count =0
for k in fl[1:]:
    k = k.split()
    pid = k[0]+'_'+k[1]
    cur.execute("select alphaHelix,patern from PDB_proteinChain WHERE pdbId_C LIKE '%s'"%pid)
    out = cur.fetchall()
    if len(out) and out[0][0] <= 10:
        patc = str(out[0][0])+'.P'
        print patc,
        cur.execute("select xf from patern_index WHERE idpatern_index='%s'"%patc)
        xf = cur.fetchall()[0][0]
        cur.execute("select xf from patern_index WHERE patern='%s'"%out[0][1])
        xo = cur.fetchall()[0][0]
        #print xf,xo
        frqType = ''
        if xo >= xf:
            frqType= "P"
        else:
            frqType= "nP"
        d1_Data.append([k[0],k[1],out[0][0],out[0][1],frqType,int(k[2]),int(k[3]),int(k[4]),float(k[5]),float(k[6]),float(k[7]),float(k[8]),float(k[9]),float(k[10]),float(k[11]),float(k[12]),float(k[13]),float(k[14])])
        count += 1
    else:
        print out
#print count
```

```python
colnames = ['pid','chain','H','Patern','frqType','length','aCont','rCont','rco','aco','sro','lro','sr_per','mr_per','lr_per','lco','nlco','tcd']
d1_DF = pd.DataFrame(d1_Data,columns=colnames)
```

```python
d1_DF.describe()
```

```python
aco_h4p = d1_DF[(d1_DF.frqType=="P") & (d1_DF.H == 4)]["aco"]
aco_h4np = d1_DF[(d1_DF.frqType=="nP") & (d1_DF.H == 4)]["aco"]

nlco_h4p = d1_DF[(d1_DF.frqType=="P") & (d1_DF.H == 4)]["nlco"]
nlco_h4np = d1_DF[(d1_DF.frqType=="nP") & (d1_DF.H == 4)]["nlco"]

lco_h4p = d1_DF[(d1_DF.frqType=="P") & (d1_DF.H == 4)]["lco"]
lco_h4np = d1_DF[(d1_DF.frqType=="nP") & (d1_DF.H == 4)]["lco"]
```

```python
import matplotlib.pyplot as plt
#%matplotlib inline
plt.figure(figsize=(10,5))#,dpi=300)

ax1 = plt.subplot(1,3,1)
ax1.boxplot((aco_h4p,aco_h4np),(1,2))
ax1.set_ylabel("Absolute Contact Order")
ax1.set_xticklabels((1,2),('P','NP'))

ax2 = plt.subplot(1,3,2)
ax2.boxplot((nlco_h4p,nlco_h4np),(1,2))
ax2.set_ylabel("non-Local Contact Order")

ax2 = plt.subplot(1,3,3)
ax2.boxplot((lco_h4p,lco_h4np),(1,2))
ax2.set_ylabel("nl -CO")

plt.suptitle("Variants of Contact Orders (4HS proteins %d)"%len(d1_DF[d1_DF.H==4]),fontsize=12)
plt.tight_layout()
plt.show()
```
